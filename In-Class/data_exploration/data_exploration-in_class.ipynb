### Data Visualization (Subplots) & Analysis

```python
#TODO: subplots for Inflation Rate & GDP
# st_drp[['Inflation rate', 'GDP']].hist(figsize=(10, 5))
st_drp[['Inflation rate', 'GDP']].hist(bins=3, figsize=(10, 5))
plt.show()
```

```python
import numpy as np
hist_values_GDP, bin_edges_GDP = np.histogram(st_drp['GDP'], bins=3)
hist_values_inflation, bin_edges_inflation = np.histogram(st_drp['Inflation rate'], bins=3)

# bin_edges_GDP.tolist()
# bin_edges_inflation.tolist()

binned_data_GDP = pd.cut(st_drp['GDP'], bin_edges_GDP.tolist())
binned_data_Inflation_Rate = pd.cut(st_drp['Inflation rate'], bin_edges_inflation.tolist())

binned_df = st_drp.copy()
binned_df['GDP_Binned'] = binned_data_GDP
binned_df['InflationRate_Binned'] = binned_data_Inflation_Rate

bin_features = ['InflationRate_Binned', 'GDP_Binned']
fig = plt.subplots(figsize=(15, 30))
for i, j in enumerate(bin_features):
    plt.subplot(4, 2, i+1)
    plt.subplots_adjust(hspace=1.0)
    sns.countplot(x=j, data=binned_df, hue='Target')
```

**Analysis**  
To see the relationship of continuous variables such as GDP and Inflation Rate to the categorical target it might be useful to bin them for visual purposes. Based on the Histograms it seems that roughly three bins for both variables seem to be good. These bins could be interpreted as "Low", "Moderate", and "High" Inflation Rate or GDP respectively.  

It can be seen from the count plots that a lots of students tend to drop when Inflation rate is very low or very high but not as much when it is moderate. Interestingly more students tend to drop during moderate and high GDP compared to lower GDP, these features could perhaps be looked at later in a cluster analysis to identify some pattern.  

Most students tend to graduate or stay enrolled when inflation is very low and GDP is very high.

```python
features = ['Gender','Educational special needs','Target', 'Scholarship holder', 'International','Displaced']

fig = plt.subplots(figsize=(15, 20))
for i, j in enumerate(features):
    plt.subplot(4, 2, i+1)
    plt.subplots_adjust(hspace=1.0)
    sns.countplot(x=j, data=st_drp, hue='Target')
    plt.xticks(rotation=90)
```

### Cluster Analysis

```python
from sklearn.cluster import KMeans

# Data Filtering
target_vals = ['Dropout']
clustering_features = ['Inflation rate', 'GDP']

target_st = st_drp[clustering_features][st_drp['Target'].isin(target_vals)]

kmeans = KMeans(n_clusters=3, random_state=0).fit(target_st)

target_st['label'] = kmeans.labels_

plt.scatter(target_st[clustering_features[0]], target_st[clustering_features[1]], c=target_st['label'])
plt.xlabel(clustering_features[0])
plt.ylabel(clustering_features[1])
plt.title('Clusters of students that dropped out')
plt.show()
```

**Analysis**  
A cluster analysis seems to indicate that students who drop out ted to either:
- live in a region with low GDP and high inflation rate
- live in a region with low GDP and low inflation rate
- live in a region with high GDP and low to moderate inflation rate (possibe bad match)

### Creating Dummy Variables

```python
# Creating Dummy Variable for Target
# Dummy Variable is DroputStatus, 1 if dropped out 0 if not (i.e. 0 if graduate or enrolled)

st_drp_encoded = st_drp.copy()

encoding = {'Dropout': 1, 'Graduate': 0, 'Enrolled': 0}

st_drp_encoded['Dropout'] = st_drp_encoded['Target'].map(encoding)
st_drp_encoded = st_drp_encoded.drop('Target', axis=1)

st_drp_encoded.head()
```

```python
# dataframe of features to use for model
student_dropout_df = st_drp_encoded[['Gender','Educational special needs', 'Scholarship holder', 'International','Displaced', 'Dropout']]
student_dropout_df.info()
```

### Fitting a Model

Splitting data into training and testing

```python
from sklearn.model_selection import train_test_split

X = student_dropout_df.drop(['Dropout'], axis=1)
y = student_dropout_df['Dropout']

# 70% Training and 30% Testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# from sklearn import utils
# print(utils.multiclass.type_of_target(y))

student_dropout_df['Dropout']
```

```python
print( X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

Fitting a model

```python
from sklearn.linear_model import LogisticRegression

dropout_logRegr = LogisticRegression()
dropout_logRegr.fit(X_train, y_train)

predictions = dropout_logRegr.predict(X_test)

print("Score:", dropout_logRegr.score(X_test, y_test))
```

### K Fold Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import numpy as np

model = LogisticRegression()
kfold = KFold(n_splits=10, random_state=None, shuffle=False)
scores = cross_val_score(model, X, y, cv=kfold)

print("Cross-Validation Scores:", scores)
print("\nMean CV Score:", np.mean(scores))
```
