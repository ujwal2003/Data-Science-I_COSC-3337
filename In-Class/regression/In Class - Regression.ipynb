```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Import the data using pandas and display the first 5 rows

```python
iris_df = pd.read_csv('iris.csv')
```

```python
iris_df.head()
```

## Goal 1: Predict the sepal length of an iris, given that we know the iris' petal width
#### First, let's make a scatterplot using petal_width as the x and sepal_length as the y

```python
iris_df.plot(kind = 'scatter', x = 'petal_width', y = 'sepal_length', figsize=(6,4))
```

### Train a Linear Model

**Let's set up the linear regression first, using petal_width as the x variable and sepal_length as the target variable**

```python
petal_width = iris_df['petal_width']
a = np.ones((petal_width.shape[0], 1))
X = np.hstack((a, petal_width.values.reshape(-1,1)))
y = iris_df['sepal_length']
```

**Now, we are going to train the linear model, display theta_0 and theta_1**

```python
theta = np.matmul(np.linalg.inv(np.matmul(X.T,X)), np.matmul(X.T,y))
theta
```

### Now let's graph your trained linear model

**Your regression should look like the one displayed below**

```python
iris_df.plot(kind = 'scatter', x = 'petal_width', y = 'sepal_length', figsize=(6,4))
plt.axis([0,3,4,10])
X = np.linspace(0,3)
plt.plot(X,theta[0]+theta[1]*X, "b") #plot the model here
plt.text(0.5, 8, r"$\theta_0 = $" + f'{round(theta[0], 2)}', fontsize=13, color="b") #display theta_0 here as shown below
plt.text(0.5, 7, r"$\theta_1 = $" + f'{round(theta[1], 2)}', fontsize=13, color="b") #display theta_1 here as shown below
plt.show()
```

#### Make Predictions, Let's say we know the petal_width is 1.37, what would be the iris' sepal length?

```python
example_petal_width = 1.37
predicted_sepal_length = theta[0] + theta[1] * example_petal_width
print(f'The predicted sepal_length is: {predicted_sepal_length}')
```

### Now let's plot our predicted sepal length onto our linear model

**Your linear model should look like the one displayed below**

```python
iris_df.plot(kind = 'scatter', x = 'petal_width', y = 'sepal_length', figsize=(6,4))
plt.axis([0,3,0,10])
X = np.linspace(0,3)
plt.plot(X, theta[0] + theta[1]*X, "b") #plot the model here
plt.text(0.5, 8,  r"$\theta_0 = $" + f'{round(theta[0], 2)}', fontsize=13, color="b") #display theta_0 here as shown below
plt.text(0.5, 7, r"$\theta_1 = $" + f'{round(theta[1], 2)}', fontsize=13, color="b") #display theta_1 here as shown below
plt.plot([example_petal_width,example_petal_width],[0,predicted_sepal_length],"r--")
plt.text(1.5, 4, f'Prediction = {round(predicted_sepal_length, 2)}', fontsize=13, color="b") #display your prediction here as shown below
plt.plot(example_petal_width, predicted_sepal_length,"ro") #plot the petal width with respect to its predicted sepal length
plt.show()
```
